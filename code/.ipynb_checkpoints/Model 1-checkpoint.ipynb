```python
import numpy as np
import tensorflow as tf
import pandas as pd

from util import load_pickle, train
```

```python
vec1 = load_pickle('column1_vec_representations.pkl')
```

```python
vec2 = load_pickle('column2_vec_representations.pkl')
```

```python
train_labels = load_pickle('labels.pkl')
```

```python
train_data = []
for v, w in zip(vec1, vec2):
    try:
        mean_vec_1 = np.mean(v, axis=0)
        if np.isnan(mean_vec_1).any():
            mean_vec_1 = np.zeros(100)
        mean_vec_2 = np.mean(w, axis=0)
        if np.isnan(mean_vec_2).any():
            mean_vec_2 = np.zeros(100)
        train_data.append(np.concatenate((mean_vec_1, mean_vec_2)))
    except RuntimeWarning:
        print v.shape,w.shape
```

```python
train_data = np.array(train_data)
```

```python
train(train_data, train_labels, 500, 20000)
```
